# **Brightness**

```python
import cv2 as cv
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# ImageDataGenerator class
datagen = ImageDataGenerator(brightness_range=[0.2,1.0])

#loop over the data in batches and this automatically saves the images
i = 0
for batch in datagen.flow_from_directory('/content/drive/MyDrive/Colab Notebooks/Master Thesis/Test3', batch_size=9,target_size=(720,960),
                          save_to_dir='/content/drive/MyDrive/Colab Notebooks/Master Thesis/Ultrasound_fetus/Augmented Images/CrossAugmentation_Brightness', save_format='jpg'):
    i += 1
    if i > 45:
        break
```

```python
import cv2
import os

def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
        if img is not None:
            images.append(img)
    return images
```

```python
images_br = load_images_from_folder('/content/drive/MyDrive/Colab Notebooks/Master Thesis/Ultrasound_fetus/Augmented Images/CrossAugmentation_Brightness')
len(images_br)
```
